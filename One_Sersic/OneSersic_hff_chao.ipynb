```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.visualization import simple_norm
import pyregion
from astropy.nddata import Cutout2D
from astropy.wcs import WCS
```

```python
hdr=fits.open('../../../../f160w/gala2_detection/f160w_icl_bcg_out_detection.fits')[0].header
exptime=hdr['exptime']

plt.figure(figsize =(2*2,3*2))
```

```python
ax1=plt.subplot(231)
data=fits.open('../../../../f160w/gala2_detection/f160w_icl_bcg_out_detection_corrected.fits')[0].data
norm=simple_norm(data,stretch='asinh',min_cut=-0.007, max_cut=0.1)
ax1.imshow(data/exptime,cmap='gray_r',norm=norm,origin='lower')
ax1.text(100,2650,'Refined processed image',fontsize=8)
ax1.get_xaxis().set_visible(False)
ax1.get_yaxis().set_visible(False)

ax2=plt.subplot(232)
data=fits.open('../../../../f160w/gala2_detection/all_model.fits')[0].data
norm=simple_norm(data,stretch='asinh',min_cut=-0.007, max_cut=0.1)
ax2.imshow(data/exptime,cmap='gray_r',norm=norm,origin='lower')
ax2.text(100,2650,r'One S$\'{\rm e}$rsic models',fontsize=8)
ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)

ax3=plt.subplot(233)
data=fits.open('../../../../f160w/gala2_detection/cluster_model.fits')[0].data
norm=simple_norm(data,stretch='asinh',min_cut=-0.007, max_cut=0.1)
ax3.imshow(data/exptime,cmap='gray_r',norm=norm,origin='lower')
ax3.text(100, 2650, 'Full models', fontsize=8)
ax3.get_xaxis().set_visible(False)
ax3.get_yaxis().set_visible(False)

ax4=plt.subplot(234)
data=fits.open('../../../../f160w/gala2_detection/f160w_2fcg_out.fits')[0].data
norm=simple_norm(data,stretch='asinh',min_cut=-0.007, max_cut=0.1)
ax4.imshow(data/exptime,cmap='gray_r',norm=norm,origin='lower')
ax4.text(100, 2650, 'Pure bCGs+ICL', fontsize=8)
ax4.get_xaxis().set_visible(False)
ax4.get_yaxis().set_visible(False)

ax5=plt.subplot(235)
data=fits.open('../../../../f160w/icl_bcg/all_model.fits')[0].data
norm=simple_norm(data,stretch='asinh',min_cut=-0.007, max_cut=0.1)
ax5.imshow(data/exptime,cmap='gray_r',norm=norm,origin='lower')
ax5.text(100, 2650, 'bCG models', fontsize=8)
ax5.get_xaxis().set_visible(False)
ax5.get_yaxis().set_visible(False)

ax6=plt.subplot(236)
data=fits.open('../../../../f160w/gala2_detection/f160w.fits')[0].data
norm=simple_norm(data,stretch='asinh',min_cut=-0.007, max_cut=0.1)
ax6.imshow(data/exptime,cmap='gray_r',norm=norm,origin='lower')
ax6.text(100, 2650, 'Original image', fontsize=8)
ax6.get_xaxis().set_visible(False)
ax6.get_yaxis().set_visible(False)

plt.tight_layout()
plt.tight_layout(w_pad=0.0001,h_pad=0.0001)
plt.savefig('catalog_fig.jpg',dpi=200)
```
